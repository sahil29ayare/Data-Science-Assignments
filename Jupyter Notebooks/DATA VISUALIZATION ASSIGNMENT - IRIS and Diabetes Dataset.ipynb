# IRIS Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_data = pd.read_csv(r"C:\Users\sahil\Downloads\archive3\Iris.csv")
iris_data.head()
```

```python
# What is the distribution of sepal lengths and widths among the Iris species?

# Visualizing the distribution of sepal lengths and widths among Iris species
plt.figure(figsize=(12, 6))

# Histogram of sepal lengths for each species
plt.subplot(1, 2, 1)
sns.histplot(data=iris_data, x='SepalLengthCm', hue='Species')
plt.title('Distribution of Sepal Lengths among Iris Species')

# Histogram of sepal widths for each species
plt.subplot(1, 2, 2)
sns.histplot(data=iris_data, x='SepalWidthCm', hue='Species', )
plt.title('Distribution of Sepal Widths among Iris Species')

plt.show()
```

```python
# How does the petal length vary across different Iris species?

# Create a box plot for petal lengths
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=iris_data)
plt.title('Distribution of Petal Lengths among Iris Species')
plt.show()
```

```python
#Is there a relationship between sepal length and petal length

plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_data, hue='Species')
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

```python
#Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?

# Create a pair plot for pairwise relationships between features
sns.pairplot(iris_data, hue='Species', palette='Set1', markers=["o", "s", "D"])
plt.suptitle('Pairwise Relationships between Features', y=1.02)
plt.show()
```

# Diabetes Dataset 

```python
diabetes_df = pd.read_csv(r"C:\Users\sahil\Downloads\archive4\diabetes.csv")
diabetes_df.head()
```

```python
#What is the distribution of age and BMI (Body Mass Index) among diabetic patients?

# Visualizing the distribution of age and BMI among diabetic patients
plt.figure(figsize=(12, 6))

# Histogram of age
plt.subplot(1, 2, 1)
sns.histplot(diabetes_df['Age'], bins=10)
plt.title('Distribution of Age among Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Histogram of BMI
plt.subplot(1, 2, 2)
sns.histplot(diabetes_df['BMI'], bins=10)
plt.title('Distribution of BMI among Diabetic Patients')
plt.xlabel('BMI')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#How do blood pressure levels vary across different age groups?

# Visualizing blood pressure levels across age groups using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='BloodPressure', data=diabetes_df)
plt.title('Blood Pressure Levels across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
#Is there a correlation between BMI and Glucose levels?

# Visualizing the relationship between BMI and Glucose levels using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Glucose', data=diabetes_df)
plt.title('BMI vs Glucose Levels')
plt.xlabel('BMI (Body Mass Index)')
plt.ylabel('Glucose Levels')
plt.show()
```

```python
#Are there any observable patterns in the relationship between diabetes progression and patient age?

# Visualizing the relationship between diabetes progression and patient age using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='DiabetesPedigreeFunction', data=diabetes_df)
plt.title('DiabetesPedigreeFunction vs Patient Age')
plt.xlabel('Age')
plt.ylabel('DiabetesPedigreeFunction')
plt.show()
```

```python
#How do insulin levels vary across different age groups?

# Visualizing insulin levels across age groups using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Insulin', data=diabetes_df)
plt.title('Insulin Levels across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Insulin Levels')
plt.show()
```
