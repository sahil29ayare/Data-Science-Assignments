```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:\Ozy\Data Science\DS datasets\dataset_heart.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Count of patients
patients_count = data.shape[0]
print(f"Total number of patients: {patients_count}")
```

```python
# Count of patients with and without heart disease
heart_disease = data['heart disease'].value_counts()
print(f"Patients without heart disease: {heart_disease[1]}")
print(f"Patients with heart disease: {heart_disease[2]}")
```

```python
# Age distribution statistics
data['age'].describe()
```

```python
# Heart disease presence based on gender
data.groupby('sex')['heart disease'].value_counts()
```

```python
# Heart disease presence across age groups
plt.figure(figsize=(10,6))
sns.histplot(data=data, x='age', hue='heart disease', bins=10)
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Heart disease presence based on chest pain type
data.groupby('chest pain type')['heart disease'].value_counts()
```

```python
# Heart disease presence based on fasting blood sugar
data.groupby('fasting blood sugar')['heart disease'].value_counts()
```

```python
# Heart disease presence based on resting electrocardiographic results
data.groupby('resting electrocardiographic results')['heart disease'].value_counts()
```

```python
# Heart disease presence based on exercise induced angina	
data.groupby('exercise induced angina')['heart disease'].value_counts()
```

```python
# Heart disease presence based on ST segment	
data.groupby('ST segment')['heart disease'].value_counts()
```

```python
# Heart disease presence based on major vessels	
data.groupby('major vessels')['heart disease'].value_counts()
```

```python
# Heart disease presence based on thal	
data.groupby('thal')['heart disease'].value_counts()
```

```python
#Identify any missing or null values and devise strategies for handling them
missing_values = data.isnull().sum()
print(missing_values)
```

```python
#Correlation between cholesterol level and heart disease
correlation_chol = data['serum cholestoral'].corr(data['heart disease'])
print(f"Correlation between cholesterol level and heart disease presence: {correlation_chol}")
```

```python
# Visualize the relationship using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum cholestoral', y='heart disease', data=data, hue='heart disease')
plt.title('Cholesterol Level vs. Heart Disease Presence')
plt.xlabel('Cholesterol Level (mg/dl)')
plt.ylabel('Heart Disease Presence (1 = No, 2 = Yes)')
plt.show()
```

```python
#correlation between age and heart disease
correlation_age = data['age'].corr(data['heart disease'])
print(f"Correlation between age and heart disease presence: {correlation_age}")
```

```python
# Visualize the relationship using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='heart disease', data=data, hue='heart disease')
plt.title('Age vs. Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Heart Disease Presence (1 = No, 2 = Yes)')
plt.show()
```

```python
#create a risk factor based on multiple features.
data['high_chol'] = data['serum cholestoral'].apply(lambda x: 1 if x > 200 else 0)
data['high_bp'] = data['resting blood pressure'].apply(lambda x: 1 if x > 120 else 0)

data['risk_factor'] = data['high_chol'] + data['high_bp']

# Visualizing the new risk factor
plt.figure(figsize=(10, 6))
sns.countplot(x='risk_factor', data=data, hue='heart disease')
plt.title('Heart Disease Presence Based on Risk Factor')
plt.xlabel('Risk Factor (High Cholesterol + High Blood Pressure)')
plt.ylabel('Count')
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = data.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
