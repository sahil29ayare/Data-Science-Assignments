```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loads the dataset from the CSV file.
data = pd.read_csv(r"C:\Ozy\Data Science\DS datasets\brca.csv")
```

```python
# Displays first 5 rows of the dataset.
data.head() 
```

```python
# Provides basic information about the dataset, including the number of non-null entries.
data.info()
```

```python
# Summary statistics for numerical columns.
data.describe(include= 'all')
```

```python
# Display the distribution of the 'diagnosis' column
diagnosis_distribution = data['diagnosis'].value_counts()
diagnosis_distribution
```

```python
# Group data by 'diagnosis' and compute summary statistics
grouped_stats = data.groupby('diagnosis').mean()

grouped_stats
```

```python
# List of key features to visualize
key_features = ['x.radius_mean', 'x.texture_mean', 'x.perimeter_mean', 'x.area_mean', 'x.smoothness_mean']

# Create a pair plot to visualize differences
sns.pairplot(data, hue="diagnosis", vars=key_features, palette="coolwarm")
plt.show()
```

- Pair Plot: This creates a pair plot to visualize the relationships between the selected key features for malignant and benign diagnoses. It helps in understanding how these features differ between the two diagnosis groups.

```python
# Create box plots for each key feature
for feature in key_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="diagnosis", y=feature, data=data, palette="coolwarm")
    plt.title(f'Distribution of {feature} by Diagnosis')
    plt.show()
```

- Box Plots: This creates box plots for each key feature, showing the distribution of the feature values for each diagnosis group. Box plots are useful for visualizing the spread and central tendency of the data, as well as identifying outliers.

```python
# Identify and handle missing or null values
data.isnull().sum() 
```

- No missing values present in the dataset

```python
# Encode the target variable 'diagnosis' to numeric values (0 for benign, 1 for malignant)
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

data.head()
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Display the correlation matrix
print(correlation_matrix['diagnosis'].sort_values(ascending=False))
data.head()
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

- The correlation matrix helps us identify features that are strongly correlated with the target variable (diagnosis).

```python
# Visualize top correlated features with the target variable
top_correlated_features = correlation_matrix['diagnosis'].abs().sort_values(ascending=False).index[1:6]
```

```python
# Create scatter plots for each of the top correlated features

for feature in top_correlated_features:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=feature, y='diagnosis', hue='diagnosis', palette="coolwarm", alpha=0.7)
    plt.title(f'Relationship between {feature} and Diagnosis')
    plt.show()
```

- Scatter plots and pair plots are used to visualize the relationships between the top correlated features and the target variable.

```python
# Standardize the features using StandardScaler from scikit-learn.(excluding the target variable diagnosis)
from sklearn.preprocessing import StandardScaler

features = data.columns[:-1]
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

print(data.head())
```

- StandardScaler is a class from the sklearn.preprocessing module. It standardizes features by removing the mean and scaling to unit variance. The result is a new DataFrame where each feature has been standardized. The mean of each column will be 0, and the standard deviation will be 1. This standardized data replaces the original data in the DataFrame for those feature columns.

```python
# Create violin plots for top correlated features
for feature in top_correlated_features:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='diagnosis', y=feature, data=data, palette="coolwarm")
    plt.title(f'Distribution of {feature} by Diagnosis')
    plt.show()
```

- Violin plots are similar to box plots but include a rotated kernel density plot on each side. They provide a more detailed view of the distribution of data.

- Violin plots can help visualize the distribution of each feature across different diagnosis classes more effectively compared to box plots
