## Filter Data by Innings

```python
# Filter for 1st innings
first_innings = df_deliveries[df_deliveries['inning'] == 1].copy()

# Filter for 2nd innings
second_innings = df_deliveries[df_deliveries['inning'] == 2].copy()
```

- We filter the df_deliveries DataFrame to separate data for 1st innings (first_innings) and 2nd innings (second_innings). Ensure .copy() is used to avoid any inadvertent changes to the original DataFrame.

## Group and Aggregate Data for 1st Innings

```python
# Group by batting team and calculate total runs scored
first_innings_runs = first_innings.groupby('batting_team')['total_runs'].sum().reset_index()

# Group by bowling team and calculate total wickets taken
first_innings_wickets = first_innings[first_innings['is_wicket'] == 1].groupby('bowling_team').size().reset_index(name='wickets_taken')

# Merge runs and wickets data for 1st innings
first_innings_stats = pd.merge(first_innings_runs, first_innings_wickets, left_on='batting_team', right_on='bowling_team', how='outer').fillna(0)
first_innings_stats.rename(columns={'batting_team': 'team'}, inplace=True)
```

## Group and Aggregate Data for 2nd Innings

```python
# Group by batting team and calculate total runs scored
second_innings_runs = second_innings.groupby('batting_team')['total_runs'].sum().reset_index()

# Group by bowling team and calculate total wickets taken
second_innings_wickets = second_innings[second_innings['is_wicket'] == 1].groupby('bowling_team').size().reset_index(name='wickets_taken')

# Merge runs and wickets data for 2nd innings
second_innings_stats = pd.merge(second_innings_runs, second_innings_wickets, left_on='batting_team', right_on='bowling_team', how='outer').fillna(0)
second_innings_stats.rename(columns={'batting_team': 'team'}, inplace=True)
```

For each innings, we aggregate data:
- Total runs scored (total_runs) by each batting team.
- Total wickets taken (wickets_taken) by each bowling team.
- We merge these aggregates to create comprehensive statistics for each team in both innings.

## Plotting 1st innings comparison

```python
plt.figure(figsize=(12, 6))
plt.bar(first_innings_stats['team'].astype(str), first_innings_stats['total_runs'], label='Runs Scored')
plt.bar(first_innings_stats['team'].astype(str), first_innings_stats['wickets_taken'], label='Wickets Taken')
plt.xlabel('Teams')
plt.ylabel('Count')
plt.title('Innings-wise Comparison: 1st Innings')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## Plotting 2nd innings comparison

```python
plt.figure(figsize=(12, 6))
plt.bar(second_innings_stats['team'], second_innings_stats['total_runs'], label='Runs Scored')
plt.bar(second_innings_stats['team'], second_innings_stats['wickets_taken'], label='Wickets Taken')
plt.xlabel('Teams')
plt.ylabel('Count')
plt.title('Innings-wise Comparison: 2nd Innings')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

- Bar plots comparing runs scored and wickets taken by teams in both 1st and 2nd innings.

## Highest run scored by a team in a single match

```python
# Group by match id and batting team, then sum the total runs
team_runs_per_match = df_deliveries.groupby(['id', 'batting_team'])['total_runs'].sum().reset_index()

# Find the maximum runs scored by a team in a single match
max_runs = team_runs_per_match.groupby('batting_team')['total_runs'].max().max()

# Find the details of the match where this maximum runs were scored
max_runs_match = team_runs_per_match.loc[team_runs_per_match['total_runs'].idxmax()]

# Print the results
print(f"Highest runs scored by a team in a single match: {max_runs}")
print(f"Details of the match:")
print(f"Match ID: {max_runs_match['id']}")
print(f"Team: {max_runs_match['batting_team']}")
print(f"Total runs scored: {max_runs_match['total_runs']}")
```

- Grouping and Summation: groupby(['id', 'batting_team'])['total_runs'].sum() groups the data by match ID and batting team, then sums up the total_runs for each combination.

- Finding Maximum: max() function is used twice (max().max()) to find the highest total runs scored by any team in any match.

- Details of the Match: idxmax() finds the index of the row with the maximum total runs, and loc[] retrieves the details (match ID, team, and total runs) for that match.
